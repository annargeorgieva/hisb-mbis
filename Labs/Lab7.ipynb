Използват се библиотеките pandas, seaborn и scikit-learn за обработка, визуализация и класификация на данни. Кода чете данни от CSV файлове и създава класификационни модели, използвайки метода на случайните гори. Точността на моделите се оценява чрез изчисляване на резултатите за обучаващите и тестовите данни. Показва се значимостта на характеристиките във всяка модел, като се изчислява и сортира важността на характеристиките.

```python
import pandas as pd
import seaborn as sb
from sklearn import tree
```

```python
exam = pd.read_csv('sample_data/exams.csv')
hd = pd.read_csv('sample_data/Heart_disease_details.csv')
```

```python
exam.head()
```

```python
X = pd.get_dummies(exam.drop('parental level of education',axis=1))
y = exam['parental level of education']
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_ts, y_tr, y_ts = train_test_split(X,y,test_size=.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, max_depth=50)
rf.fit(X_tr,y_tr)
```

```python
rf.score(X_tr,y_tr), rf.score(X_ts,y_ts)
```

```python
importances = rf.feature_importances_
pd.Series(importances, index = X.columns)
```

```python
X = pd.get_dummies(hd.drop('Diagnosis',axis=1))
y = hd['Diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.25)
rf2 = RandomForestClassifier(n_estimators=100, max_depth=50)
rf2.fit(X_train,y_train)
```

```python
rf2.score(X_train,y_train), rf2.score(X_test,y_test)
```

```python
importances = rf2.feature_importances_
pd.Series(importances, index = X.columns).sort_values(ascending=False)[:40]
```
