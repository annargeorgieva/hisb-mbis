```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
import pandas as ps
```

```python
hd = ps.read_csv('sample_data/Heart_disease_details.csv')
hd.head
```

```python
plt.hist(hd['Cholesterol level (mg/dL)'])
```

```python
plt.hist('Systolic', bins=20, range=(100, 170), color='Green', alpha=1, data=hd)
plt.hist('Diastolic', bins=20, range=(50, 130), color='Red', alpha=.7, data=hd)
```

```python
plt.pie(hd['Diagnosis'].value_counts(), labels=hd.Diagnosis.value_counts().keys())
plt.title('Diffrent Diagnosis')
```

```python
plt.pie(hd.Gender.value_counts(), labels=hd.Gender.value_counts().keys())
plt.title('Gender distribution')
```

```python
colouring={'Male':'#0047AB', 'Female':'Pink'}
plt.scatter('Diastolic','Systolic', c=hd.Gender.map(colouring), data=hd)
```
